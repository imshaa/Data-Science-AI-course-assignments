```python
# WEEK 3 - DATA VISUALIZATION
# Dataset: students.csv

# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 2: Load dataset
df = pd.read_csv('/content/students.csv', sep=';')

# ------------------------------------------------
# CLASS TASK
# (Bar Chart, Scatter Plot, Histogram)
# ------------------------------------------------

# 1. Bar Chart: Average Grades by Gender
print("\nCLASS TASK (BAR CHART, SCATTER PLOT, HISTOGRAM) : ")
print("\nBAR CHART : \n")
avg_scores = df.groupby('sex')[['G1', 'G2', 'G3']].mean()
avg_scores.plot(kind='bar', figsize=(6,4))
plt.title('Average Grades by Gender')
plt.xlabel('Gender (F = Female, M = Male)')
plt.ylabel('Average Grade')
plt.show()
# Insight: Female students slightly outperform males in average grades.

# 2. Scatter Plot: Study Time vs Final Grade
print("\nSCATTER PLOT : \n")
plt.figure(figsize=(6,4))
sns.scatterplot(x='studytime', y='G3', hue='sex', data=df)
plt.title('Study Time vs Final Grade (G3)')
plt.xlabel('Study Time (1–4)')
plt.ylabel('Final Grade (0–20)')
plt.show()
# Insight: More study time is generally linked to higher final grades.

# 3. Histogram: Distribution of Final Grades
print("\nHISTOGRAM : \n")
plt.figure(figsize=(6,4))
plt.hist(df['G3'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Final Grades (G3)')
plt.xlabel('Final Grade')
plt.ylabel('Number of Students')
plt.show()
# Insight: Most students score between 10 and 15, showing average performance.

# ------------------------------------------------
# ASSIGNMENT 3
# (Create 5 different plots from dataset)
# ------------------------------------------------
print("\nASSIGNMENT 3 (5 DIFFERENT PLOTS) : \n")

# 4. Boxplot: Absences vs Higher Education
print("\nBOXPLOT : \n")
plt.figure(figsize=(6,4))
sns.boxplot(x='higher', y='absences', data=df)
plt.title('Absences by Desire for Higher Education')
plt.xlabel('Wants Higher Education (yes/no)')
plt.ylabel('Number of Absences')
plt.show()
# Insight: Students wanting higher education usually have fewer absences.

# 5. Count Plot: Mother's Education vs Performance
print("\nCOUNT PLOT : \n")
df['Performance'] = pd.cut(df['G3'], bins=[0,10,15,20], labels=['Low','Average','High'])
plt.figure(figsize=(6,4))
sns.countplot(x='Medu', hue='Performance', data=df)
plt.title("Mother's Education Level vs Student Performance")
plt.xlabel("Mother's Education (0–4)")
plt.ylabel("Number of Students")
plt.show()
# Insight: Students with more educated mothers tend to perform better.

# --- Plot 6: Heatmap - Correlation Between Key Numeric Features ---
print("\nHEATMAP : \n")
plt.figure(figsize=(7,5))
sns.heatmap(df[['G1','G2','G3','studytime','absences']].corr(), annot=True, cmap='Blues')
plt.title('Correlation Between Academic Features')
plt.show()
# Insight: G1 and G2 have the strongest correlation with final grade (G3).

# --- Plot 7: Pie Chart - Students Who Want Higher Education ---
print("\nPIE CHART : \n")
plt.figure(figsize=(5,5))
df['higher'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=['lightblue','lightgreen'])
plt.title('Students Who Want Higher Education')
plt.ylabel('')
plt.show()
# Insight: Majority of students aim to pursue higher education.

# --- Plot 8: Line Graph - Average Final Grade by Study Time ---
print("\nLINE GRAPH : \n")
avg_study = df.groupby('studytime')['G3'].mean()
plt.figure(figsize=(6,4))
plt.plot(avg_study.index, avg_study.values, marker='o', linestyle='-', color='orange')
plt.title('Average Final Grade by Study Time')
plt.xlabel('Study Time (1–4)')
plt.ylabel('Average Final Grade (G3)')
plt.show()
# Insight: Higher study time levels show an upward trend in average grades.
```
