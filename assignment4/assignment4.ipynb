```python
# ==========================================================
# DATA SCIENCE ASSIGNMENT - WEEK 4
# Topic: Statistics & Probability
# ==========================================================
# CLASS TASK:
#   - Calculate Mean, Median, Mode, Variance, and Correlation
#
# ASSIGNMENT 4:
#   - Perform Correlation Analysis in your dataset
#   - Identify which 3 features are most related to target variable (G3)
# ==========================================================

# Step 1: Import necessary libraries
import pandas as pd
import numpy as np
from scipy.stats import pearsonr

# Step 2: Load dataset
df = pd.read_csv('/content/students.csv', sep=';')



# Step 4: Select numeric columns for statistical analysis
cols = ['G1', 'G2', 'G3', 'studytime', 'absences', 'famrel', 'freetime']

# Step 5: Calculate Mean, Median, Mode, Variance
print("===== CLASS TASK: BASIC STATISTICS =====\n")
for c in cols:
    data = df[c].dropna()
    mean_val = round(data.mean(), 2)
    median_val = round(data.median(), 2)
    mode_val = data.mode()[0] if not data.mode().empty else np.nan
    variance_val = round(data.var(ddof=0), 2)

    print(f"{c}:")
    print(f"  Mean     = {mean_val}")
    print(f"  Median   = {median_val}")
    print(f"  Mode     = {mode_val}")
    print(f"  Variance = {variance_val}\n")

# Step 6: Correlation Matrix
print("===== ASSIGNMENT 4: CORRELATION ANALYSIS =====\n")
corr = df[cols].corr()
print("Correlation Matrix (numeric columns):\n")
print(corr, "\n")

# Step 7: Correlation of each feature with final grade (G3)
target = 'G3'
corr_with_g3 = corr[target].drop(target).sort_values(key=lambda x: abs(x), ascending=False)

print("Correlation of each feature with G3 (Final Grade):\n")
print(corr_with_g3, "\n")

# Step 8: Identify Top 3 Most Related Features
top3 = corr_with_g3.head(3)
print("===== TOP 3 FEATURES MOST RELATED TO G3 =====\n")
print(top3, "\n")


print("\n===== REPORT SUMMARY =====")
print("- Mean, median, mode, and variance describe data distribution for each feature.")
print("- Correlation shows how features are related to the final grade (G3).")
print("- Top 3 most related features indicate which ones help predict student performance.")
```
